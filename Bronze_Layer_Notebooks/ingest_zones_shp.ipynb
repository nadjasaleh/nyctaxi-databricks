```python
import geopandas as gpd
```

```python
path = "/Volumes/group3_gp/adlslanding/geodata_test/taxi_zones.shp"
gdf = gpd.read_file(path)
```

```python
gdf["geometry_wkt"] = gdf.geometry.to_wkt()
gdf = gdf.drop(columns="geometry")
```

```python
spark_df = spark.createDataFrame(gdf)
spark_df.write.format("delta").mode("overwrite").saveAsTable("bronze.taxi_zones")
```
