```python
# Testing of Fact table creation
df = spark.sql("""
    SELECT
        d.Year,
        COUNT(*) AS total_trips,
        SUM(t.trip_distance) AS total_trip_distance,
        SUM(t.trip_time) AS total_trip_time,
        SUM(t.total_amount) AS total_revenue
    FROM group3_gp.silver.combined_taxi_trips t
    JOIN group3_gp.gold.dim_date d
        ON year(t.pickup_datetime) = d.year
    GROUP BY d.Year
    ORDER BY d.Year
""")

df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("group3_gp.gold.Fact_test")
```
